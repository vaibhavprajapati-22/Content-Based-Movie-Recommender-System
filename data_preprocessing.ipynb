```python
# Importing pandas library
import pandas as pd
```

```python
#Loading the dataset
columns_movies = ['ID', 'Title', 'Genres']
movies = pd.read_csv('movies.dat', sep='::', engine='python', header=None, names=columns_movies)
movies['Genres'] = movies['Genres'].str.replace('|', ', ')
movies['Info'] = movies['Title'] +" " + movies['Genres']
movies
```

```python
# Extracting the names of Movies and storing it in a list
all_movies_names = movies['Title']
len(all_movies_names)
```

```python
# # Writing each movie name to a new line in the 'movies.txt' file
with open('movies.txt', 'w', encoding='utf-8') as output_file:
    for movie_name in all_movies_names:
        output_file.write(movie_name + '\n')
```
